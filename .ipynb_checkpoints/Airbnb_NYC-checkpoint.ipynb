## Analise de Percentis de Preço por Vizinhança

```python
# Percentis dos preços por noite por cada vizinhança/zona

# Array das zonas existentes no dataset
vizinhanca_grupos = ['Brooklyn', 'Queens', 'Bronx', 'Manhattan', 'Staten Island']

#Brooklyn
zona_1 = df.loc[df['neighbourhood_group'] == vizinhanca_grupos[0]]
zona_1 = zona_1.loc[zona_1['price'] > 0] # queremos que o preço seja valido e maior que zero para a nossa estatistica
precos_zona_1 = zona_1[['price']]

#Queens
zona_2 = df.loc[df['neighbourhood_group'] == vizinhanca_grupos[1]]
zona_2 = zona_2.loc[zona_2['price'] > 0] # queremos que o preço seja valido e maior que zero para a nossa estatistica
precos_zona_2 = zona_2[['price']]

#Bronx
zona_3 = df.loc[df['neighbourhood_group'] ==  vizinhanca_grupos[2]]
zona_3 = zona_3.loc[zona_3['price'] > 0] # queremos que o preço seja valido e maior que zero para a nossa estatistica
precos_zona_3 = zona_3[['price']]

#Manhattan
zona_4 = df.loc[df['neighbourhood_group'] ==  vizinhanca_grupos[3]]
zona_4 = zona_4.loc[zona_4['price'] > 0] # queremos que o preço seja valido e maior que zero para a nossa estatistica
precos_zona_4 = zona_4[['price']]

#Staten Island
zona_5 = df.loc[df['neighbourhood_group'] == vizinhanca_grupos[4]]
zona_5 = zona_5.loc[zona_5['price'] > 0] # queremos que o preço seja valido e maior que zero para a nossa estatistica
precos_zona_5 = zona_5[['price']]

# Agrupar os preços por vizinhança/zona num array
lista_de_precos = [precos_zona_1, precos_zona_2, precos_zona_3, precos_zona_4, precos_zona_5]

percentagem_precos_por_zona = []

#creating a for loop to get statistics for price ranges and append it to our empty list
for precos_zona_x in lista_de_precos:
    # Descartar o count, mean e std (3:) e dscrever os preços desta zona por percentis 

    percentis = precos_zona_x.describe(percentiles=[.25, .50, .75]).iloc[3:]
    
    # Indexar os dados
    percentis.reset_index(inplace=True)
    
    # Renomear o index para Percentagem
    percentis.rename(columns={'index':'Percentagem'}, inplace=True)
    
    # Adicionar os dados (price e Stats) da zona processados no array
    percentagem_precos_por_zona.append(percentis)
     
# Renomear o nome da coluna price para a zona/vizinhança em analise   
percentagem_precos_por_zona[0].rename(columns={'price':vizinhanca_grupos[0]}, inplace=True)
percentagem_precos_por_zona[1].rename(columns={'price':vizinhanca_grupos[1]}, inplace=True)
percentagem_precos_por_zona[2].rename(columns={'price':vizinhanca_grupos[2]}, inplace=True)
percentagem_precos_por_zona[3].rename(columns={'price':vizinhanca_grupos[3]}, inplace=True)
percentagem_precos_por_zona[4].rename(columns={'price':vizinhanca_grupos[4]}, inplace=True)

# Preparar os dados para a vizualicação 
stat_df = [df.set_index('Percentagem') for df in percentagem_precos_por_zona]
stat_df = stat_df[0].join(stat_df[1:])
stat_df
```

## Analise de Medias de Preço por Vizinhança

```python
# Media dos preços por noite por cada vizinhança/zona

# Array das zonas existentes no dataset
vizinhanca_grupos = ['Brooklyn', 'Queens', 'Bronx', 'Manhattan', 'Staten Island']

#Brooklyn
zona_1 = df.loc[df['neighbourhood_group'] == vizinhanca_grupos[0]]
zona_1 = zona_1.loc[zona_1['price'] > 0] # queremos que o preço seja valido e maior que zero para a nossa estatistica
precos_zona_1 = zona_1[['price']]

#Queens
zona_2 = df.loc[df['neighbourhood_group'] == vizinhanca_grupos[1]]
zona_2 = zona_2.loc[zona_2['price'] > 0] # queremos que o preço seja valido e maior que zero para a nossa estatistica
precos_zona_2 = zona_2[['price']]

#Bronx
zona_3 = df.loc[df['neighbourhood_group'] ==  vizinhanca_grupos[2]]
zona_3 = zona_3.loc[zona_3['price'] > 0] # queremos que o preço seja valido e maior que zero para a nossa estatistica
precos_zona_3 = zona_3[['price']]

#Manhattan
zona_4 = df.loc[df['neighbourhood_group'] ==  vizinhanca_grupos[3]]
zona_4 = zona_4.loc[zona_4['price'] > 0] # queremos que o preço seja valido e maior que zero para a nossa estatistica
precos_zona_4 = zona_4[['price']]

#Staten Island
zona_5 = df.loc[df['neighbourhood_group'] == vizinhanca_grupos[4]]
zona_5 = zona_5.loc[zona_5['price'] > 0] # queremos que o preço seja valido e maior que zero para a nossa estatistica
precos_zona_5 = zona_5[['price']]

# Agrupar os preços por vizinhança/zona num array
lista_de_precos = [precos_zona_1, precos_zona_2, precos_zona_3, precos_zona_4, precos_zona_5]

media_precos_por_zona = []

#creating a for loop to get statistics for price ranges and append it to our empty list
for precos_zona_x in lista_de_precos:
    # usamos o mean dos valores e descartamos o count e as 6 (std, min, 25, 50, 75, max)
    media = precos_zona_x.describe().iloc[1:-6]

    # Indexar os dados
    media.reset_index(inplace=True)
    
    # Renomear o index para Percentagem
    media.rename(columns={'index':'Media'}, inplace=True)
    
    # Adicionar os dados (price e Stats) da zona processados no array
    media_precos_por_zona.append(media)
    
# Renomear o nome da coluna price para a zona/vizinhança em analise   
media_precos_por_zona[0].rename(columns={'price':vizinhanca_grupos[0]}, inplace=True)
media_precos_por_zona[1].rename(columns={'price':vizinhanca_grupos[1]}, inplace=True)
media_precos_por_zona[2].rename(columns={'price':vizinhanca_grupos[2]}, inplace=True)
media_precos_por_zona[3].rename(columns={'price':vizinhanca_grupos[3]}, inplace=True)
media_precos_por_zona[4].rename(columns={'price':vizinhanca_grupos[4]}, inplace=True)

# Preparar os dados para a vizualicação 
stat_df = [df.set_index('Media') for df in media_precos_por_zona]
stat_df = stat_df[0].join(stat_df[1:])
stat_df
```

## Regresão e Predição de Preços dos quartos (utilizando o modelo Scikit)

```python
# preprocessing.LabelEncoder() permite encodificar dados por labels de valores entre 0 e n_classes-1

encode = preprocessing.LabelEncoder() # aplica um encoder que permite normalizar labels 
encode.fit(df.neighbourhood_group) # faz fit do label ao encoder
df.neighbourhood_group = encode.transform(df.neighbourhood_group) # transformar o label num enconding normalizado

encode = preprocessing.LabelEncoder() # aplica um encoder que permite normalizar labels 
encode.fit(df.neighbourhood) # faz fit do label ao encoder
df.neighbourhood = encode.transform(df.neighbourhood) # transformar o label num enconding normalizado

encode = preprocessing.LabelEncoder() # aplica um encoder que permite normalizar labels 
encode.fit(df.room_type) # faz fit do label ao encoder
df.room_type = encode.transform(df.room_type)# transformar o label num enconding normalizado

df.sort_values(by = 'price', ascending = True, inplace = True) # ordenar os preços do quarto de modo ascendente


# Definicão do x e y para ser usado no train test split

x = df[['id','host_id','neighbourhood_group','neighbourhood','latitude','longitude','room_type','minimum_nights','number_of_reviews','availability_365']]
y = df['price']

# Divide arrays ou matrizes em numeros random de subsets de trains/tests
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=100)

# Computar Regressão Linear
regressao_linear = LinearRegression().fit(x_train, y_train)

predicao = regressao_linear.predict(x_test)
print("Erro Métrico Quadrático: ", np.sqrt(metrics.mean_squared_error(y_test, predicao)))
print("Coeficiente de determinação (R²): ", r2_score(y_test, predicao) * 100)
print("Erro Médio Absoluto: ", mean_absolute_error(y_test, predicao))
print("Raiz Quadrada do Erro Métrico Quadrático: ", mean_squared_error(y_test, predicao))

```

```python
# Criar um dataframe com os valores actuais e os valores previstos

val_actuais_previstos_df = pd.DataFrame({
'valores_actuais': np.array(y_test).flatten(),
'valores_previstos': predicao.flatten()}).head(15) # 15 elementos de comparação apenas (há mais)

width = 0.50
x = val_actuais_previstos_df.index
fig, ax = plt.subplots()
ax.bar(x + width/2, val_actuais_previstos_df.valores_actuais, width - 0.1, label='Valores Actuais')
ax.bar(x - width/2, val_actuais_previstos_df.valores_previstos, width - 0.1, label='Valores Previstos')
ax.set_ylabel('Preço')
ax.set_title('Valores Actuais vs Valores Previstos Pela Regressão Linear')
ax.legend()
plt.show()
```
